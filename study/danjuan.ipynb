```python
import numpy as np 
import pandas as pd 
import requests as req
import matplotlib.pyplot as plt 
import seaborn as sns
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.figure(figsize=(10,9))

headers={"Host":"danjuanapp.com","User-Agent":"Mozilla/5.0 (Windows NT 10.0; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/78.0.3904.108 Safari/537.36"
        }

res=req.get('https://danjuanapp.com/djapi/index_eva/dj',headers=headers)
data=res.json()
data=data["data"]["items"]
data=pd.DataFrame(data).set_index('id')
# data["pe"]=data["pe"].astype("float") 
# data.head()
data_low=data[data['pe']<7].sort_values(by='pb').iloc[:5]
data_high=data[data['pe']>30].sort_values(by='pb',ascending=False).iloc[:5]
df=data_high.append(data_low)

sns.set_context("talk", font_scale=1, rc={"lines.linewidth": 2.5})

sns.scatterplot(x=df['pe'],y=df['pb'],hue=df['name'],markers='v')
# pd.DataFrame(data,index=list(data['pe']))
```
